**We see that our features varies a lot and measured in different scales. E.g. some features have std=500 and some about 0.05. For numerical stablitity and faster convergence, we need to standardize features**


Feature standardization is a preprocessing technique used to transform features in a dataset so that they have a mean of 0 and a standard deviation of 1. This process is also known as Z-score normalization.

The formula for standardizing a single data point $x$ for a given feature is:

$$
z = \frac{x - \mu}{\sigma}
$$

Where:
*   $z$ is the standardized value.
*   $x$ is the original feature value.
*   $\mu$ (mu) is the mean of the feature values in the training data.
*   $\sigma$ (sigma) is the standard deviation of the feature values in the training data.
